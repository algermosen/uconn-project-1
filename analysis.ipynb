```python
def remove_columns_prefix(col, starting_with):
    for s in starting_with:
        replacement = col.replace(s, '')
        if replacement != col:
            return replacement.replace(' ', '_')
    return col.replace(' ', '_')
```

```python
def modify_df_column_levels(df):
    df.columns = [f"{col[0]}_{col[1]}" for col in df.columns.values]

def clean_column_names(df):
    unnamed_prefixes = [f"Unnamed: {i}_level_0_" for i in [0, 1, 2, 3]]
    unnamed_prefixes.append('Calendar Year ')
    unnamed_prefixes.append('Yearly Change  _')
    # Rename columns
    cols = df.columns.to_series().apply(lambda col: remove_columns_prefix(col, unnamed_prefixes))

    df.columns = cols

    df.rename(columns={'Medicare_Billing_Code_(HCPCS_Code)': 'HCPCS_Code'}, inplace=True)
```

```python
medications_df_list = [medications2019_df, medications2020_df, medications2021_df, medications2022_df]

[modify_df_column_levels(df) for df in medications_df_list]
[clean_column_names(df) for df in medications_df_list]
print("Done!")
```

```python
codes_2019 = set(medications2019_df['HCPCS_Code'])
codes_2020 = set(medications2020_df['HCPCS_Code'])
codes_2021 = set(medications2021_df['HCPCS_Code'])
codes_2022 = set(medications2022_df['HCPCS_Code'])

unique_codes = pd.Series(list(codes_2019.intersection(codes_2020).intersection(codes_2021).intersection(codes_2022)))
unique_codes.count()
```

```python
medications2020_df = medications2020_df.drop(columns=medications2020_df.loc[:,~medications2020_df.columns.str.startswith(tuple(['2016', 'HCPCS_Code']))].columns)
medications2022_df = medications2022_df.drop(columns=medications2022_df.loc[:,~medications2022_df.columns.str.startswith(tuple(['2022', 'HCPCS_Code']))].columns)
```

```python
medications_df = medications2021_df.merge(medications2020_df, how='inner', on='HCPCS_Code').merge(medications2022_df, how='inner', on='HCPCS_Code')
medications_df.drop(columns=medications_df.loc[:,medications_df.columns.str.startswith(tuple(['Change_in', 'Annual_Growth']))].columns, inplace=True)
medications_df
```

```python
# pd.to_numeric(medications_df[medications_df.loc[:,medications_df.columns.str.startswith('20')].columns], errors='coerce')
for col in medications_df.loc[:,medications_df.columns.str.startswith('20')].columns:
    medications_df[col] = pd.to_numeric(medications_df[col], errors='coerce').astype(int)
```

```python
medications_df.dropna(inplace=True)
```
